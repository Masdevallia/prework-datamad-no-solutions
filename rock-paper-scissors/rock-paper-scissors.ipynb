```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

from random import choice

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

options = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

maxgames = 5

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games

wingames = int(maxgames/3)+1

print("\nScissors, paper, stone: The first to win", wingames, "rounds out of", maxgames, "wins the game. Good luck!")

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

def yourchoice():
    machinechoice = choice(options)
    return machinechoice
# print(yourchoice())

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

def mychoice():
    userchoice = input("What is your choice? Stone, paper or scissors? ")
    userchoice = userchoice.lower()

    while userchoice != "stone" and userchoice != "paper" and userchoice != "scissors":
        print("""
        You must enter a correct answer!
        """)
        userchoice = input("Try again: What is your choice? Stone, paper or scissors? ")
        userchoice = userchoice.lower()
    # else:
    #     print("Your choice is", userchoice)
    return userchoice
# print(mychoice())

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

def letsplay():
    machine = yourchoice()
    me = mychoice()
    if machine == me:
        return(0)
    elif machine != me:
        if machine == "stone" and me == "paper":
            return(2)
        elif machine == "stone" and me == "scissors":
            return(1)
        elif machine == "paper" and me == "stone":
            return(1)
        elif machine == "paper" and me == "scissors":
            return(2)
        elif machine == "scissors" and me == "stone":
            return(2)
        elif machine == "scissors" and me == "paper":
            return(1)
# print(letsplay())
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated

def letsplay2():
    machine = yourchoice()
    me = mychoice()
    print(" User's choice =", me)
    print(" Machine's choice =", machine)
    if machine == me:
        print("Tie!")
        return(0)
    elif machine != me:
        if machine == "stone" and me == "paper":
            print("You win this round!")
            return(2)
        elif machine == "stone" and me == "scissors":
            print("Machine wins this round!")
            return(1)
        elif machine == "paper" and me == "stone":
            print("Machine wins this round!")
            return(1)
        elif machine == "paper" and me == "scissors":
            print("You win this round!")
            return(2)
        elif machine == "scissors" and me == "stone":
            print("You win this round!")
            return(2)
        elif machine == "scissors" and me == "paper":
            print("Machine wins this round!")
            return(1)
        
# print(letsplay2())
# letsplay2()
    
# Create two variables that accumulate the wins of each participant

MachineWins = 0
YouWin = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

# Primer intento (no tiene en cuenta el número máximo de rondas)
# while MachineWins < wingames and YouWin < wingames:
#     resultado = letsplay2()
#     if resultado == 1:
#         MachineWins += 1
#     elif resultado == 2:
#         YouWin += 1
#     print(" Machine victories:", MachineWins)
#     print(" User victories:", YouWin)

gameround = 1
while MachineWins < wingames and YouWin < wingames:
    if gameround <= maxgames: 
        print("\nRound", gameround)
        gameround += 1
        resultado = letsplay2()
        if resultado == 1:
            MachineWins += 1
        elif resultado == 2:
            YouWin += 1
        print(" Machine victories:", MachineWins)
        print(" User victories:", YouWin)
    else:
        break

# Print by console the winner of the game based on who has more accumulated wins

# if MachineWins > YouWin:
#     print("\nWinner of the game: Machine")
# else:
#     print("\nWinner of the game: User")

if MachineWins >= wingames or YouWin >= wingames:
    if MachineWins > YouWin:
        print("\nWinner of the game: Machine")
    else:
        print("\nWinner of the game: User")
else:
    print("\nYou have reached the maximum number of rounds without a winner")
```
